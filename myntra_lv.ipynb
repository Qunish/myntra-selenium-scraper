```python
def get_chrome_driver(headless_flag):
    # Set up the Selenium webdriver
    from selenium import webdriver
    from selenium.webdriver.chrome.options import Options
    from selenium.webdriver.chrome.service import Service

    service = Service(executable_path=DRIVER_FILE_PATH)
    chrome_options = Options()

    if headless_flag:
        # in case you want headless browser
        chrome_options.add_argument("--headless")
        chrome_options.add_argument("--disable-extensions")
        chrome_options.add_argument("--disable-gpu")
        chrome_options.add_argument("--no-sandbox")
        chrome_options.add_argument("--start-maximized")
        chrome_options.add_argument("user-agent={}".format(random.choice(USER_AGENT_LIST)))
    else:
        # in case  you want to open browser
        chrome_options.add_argument("--start-maximized")
        chrome_options.add_argument("user-agent={}".format(random.choice(USER_AGENT_LIST)))
        chrome_options.headless = False

    driver = webdriver.Chrome(service=service, options=chrome_options) 

    return driver
```

```python
driver = get_chrome_driver(False)
```

```python
collection_url = "https://www.myntra.com/laundry-bag?p="
```

```python
def get_card_data(card):
    try:
        price = card.find_element(By.CSS_SELECTOR,'.product-discountedPrice').text
    except Exception as e:
        price = ''
    try:
        striked_price = card.find_element(By.CSS_SELECTOR,'.product-strike').text
    except Exception as e:
        striked_price = ''
    try:
        discount = card.find_element(By.CSS_SELECTOR,'.product-discountPercentage').text
    except Exception as e:
        discount = ''
        
    try:
        product_size = card.find_element(By.CSS_SELECTOR,'.product-sizes').text
    except Exception as e:
        product_size = ''
        
    try:
        product_name = card.find_element(By.CSS_SELECTOR,'.product-productMetaInfo h3').text
    except Exception as e:
        product_name = ''
    try:
        product_desc = card.find_element(By.CSS_SELECTOR,'.product-productMetaInfo h4').text
    except Exception as e:
        product_desc = ''
    try:
        imgs = []
        images = card.find_element(By.CSS_SELECTOR,'.img-responsive source').get_attribute('srcset').split("\n")
        for item in images:
            if item !='':
                imgs.append(item.strip().split(' ')[0])
    except Exception as e:
        imgs = []
    try:
        product_url = card.find_element(By.CSS_SELECTOR,'.product-base a').get_attribute('href')
    except Exception as e:
        product_url = ''
        
    result = {}
    result['price'] = price
    result['striked_price'] = striked_price
    result['discount'] = discount
    result['product_size'] = product_size
    result['product_name'] = product_name
    result['product_desc'] = product_desc
    result['imgs'] = imgs
    result['product_url'] = product_url

    return result
```

```python
all_data = []
```

```python
page_number = 1

while True:
    try:
        current_page_url = collection_url + str(page_number)

        driver.get(current_page_url)

        # Find all cards on the current page inside the product section
        product_section = driver.find_element(By.CSS_SELECTOR, '.results-base')
        # print(product_section.text)
        all_cards = product_section.find_elements(By.CSS_SELECTOR, '.product-base')

        for card in all_cards:
            card_result = get_card_data(card)
            all_data.append(card_result)
            print(card_result)

        if not all_cards:
            break
            
        # Display the number of cards on the current page
        num_cards_on_page = len(all_cards)
        print(f"Page {page_number} - Number of cards: {num_cards_on_page}")

        page_number += 1

    except Exception as e:
        print(f"An error occurred: {e}")
        break
```
